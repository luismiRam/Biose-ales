```python
import matplotlib.pyplot as plt 
# import seaborn as sns
# sns.set()

n=np.arange(0,101,1)
y1=np.sin(np.pi*0.12*n)
plt.plot(n,y1)
```

```python
y2=np.cos(2*np.pi*0.03*n)
plt.plot(n,y2)
```

```python
y3=y1+y2
plt.plot(n,y3)
```

```python
y4=y1*y2
plt.plot(n,y4)
```

```python
n_rows = 2
n_cols = 2
listanom=[y1,y2,y3,y4]
fig, axes = plt.subplots(n_rows, n_cols)
for row_num in range(n_rows):
    for col_num in range(n_cols):
        ax = axes[row_num][col_num]
        ax.plot(n,listanom[row_num+col_num])
        ax.set_title("grafica"+str(row_num+col_num+1))
        
fig.suptitle("4 graficos")
fig.tight_layout() #Minimiza la superporsicion de graficos subplots
```

```python
n_rows = 2
n_cols = 1
listanom=[y1,y2,y3,y4]
fig, axes = plt.subplots(n_rows, n_cols)

ax = axes[0]
ax.plot(n,listanom[0],label="a")
ax.plot(n,listanom[1],label="b")
ax.set_xlabel("tiempo")
ax.set_ylabel("amplitud")
ax.set_title("grafica 1")
ax.legend()
        
ax1 = axes[1]
ax1.plot(n,listanom[2],label="c",color="red")
ax1.plot(n,listanom[3],label="d",color="green")
ax1.set_xlabel("tiempo")
ax1.set_ylabel("amplitud")
ax1.set_title("grafica 2")
ax1.legend()

fig.tight_layout()
```

```python
import pandas as pd
def reto(dic):
  arrayn=np.array([])
  for i in dic:
    arrayn=np.append(arrayn,dic[i])
  mean=np.mean(arrayn)
  max=np.max(arrayn)
  min=np.min(arrayn)
  std=np.std(arrayn)
  return(pd.Series([mean,max,min,std],index=["media","max","min","std"]))


D = {'a':1, 'b':2, 'c':3}
print(reto(D))
```
