```python
# Introducci√≥n a Python y Numpy

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a)
a=np.array([3.1,1,-0.5,-3.2,6])
b=np.array([1,3,2.2,5.1,1])
```

```python
#b) Multiplicaci√≥n escalar
producto_escalar= np.dot(a,b)
print(f"El producto escalar de los vectores a y b es: {producto_escalar}")
```

```python
# c) Multiplicacion punto a punto
producto_puntopunto= a * b
print(f"El producto punto a punto de los vectores a y b es: {producto_puntopunto}")
```

```python
# d) Construir la matriz
A= np.array([[2, -1, -3] , [4, 1.5, -2.5] , [7.3, -0.9, 0.2]])
print(f"Matriz: {A}")
```

```python
# e) Transpuesta de la matriz A
A_transpuesta= np.transpose(A)
print(f" A transpuesta : {A_transpuesta}")
```

```python
# f) 
# ones: Esta funci√≥n crea una matriz de forma especificada (tupla de enteros) y la llena con unos
matriz_ones = np.ones((2, 3)) # matriz de unos de 2x3
print(f"Matriz de unos : {matriz_ones}")

# round: Esta fnci√≥n redondea los elementos de la matriz al n√∫mero entero m√°s cercano
matriz_redondeada = np.round(A)
print(f"Matriz redondeada:{matriz_redondeada}")

# ceil:Devuelve el n√∫mero entero m√°s pr√≥ximo que sea mayor o igual a √©l.
matriz_ceil = np.ceil(A)
print(f"Matriz ceil:{matriz_ceil}")

# floor: Redondea hacia abajo cada elemento de la matriz hacia su entero m√°s cercano
matriz_floor= np.floor(A)
print(f"Matriz floor:{matriz_floor}")

# g) Acceder al valor de la primera fila y tercera columna (columna 2)
valor = A [0, 2]
print(f"El valor en la primera fila, tercera columna de la matriz A es:{valor}")
```

```python
# g) Acceder al valor de la primera fila y tercera columna (columna 2)
valor = A [0, 2]
print(f"El valor en la primera fila, tercera columna de la matriz A es:{valor}")
```

```python
# h) Obtener la segunda fila 
segunda_fila = A[1]
print(f"La segunda fila de la matriz A es:{segunda_fila}")
```

```python
# i) Dimensiones de una matriz: 
'''Se puede usar el atributo 'shape' el cual devuelve una tupla que representa las dimensiones de la 
matriz, donde el primer elemento de la tupla es el n√∫mero de filas y el segundo elemento es el n√∫mero 
de columnas.'''
dimensiones = A.shape
print(f"Dimensiones de la matriz A:{dimensiones}")
```

```python
# j) Construya la funci√≥n ùë¶[ùëõ]=sin‚Å°(ùúã‚àó0.12ùëõ) en el intervalo 0 <= n <= 100
n = np.arange(0, 101)
y = np.sin(np.pi * 0.12 * n)
```

```python
# k) Construya otra senoidal ùë¶2[ùëõ]=cos(2ùúã‚àó0.03ùëõ)
y2 = np.cos(2 * np.pi * 0.03 * n)
```

```python
# l) Genere una tercera se√±al que sea la suma de estas dos se√±ales, es decir
#    ùë†[ùëõ]=y[n]+ùë¶2[ùëõ] y una cuarta, que sea el producto de estas, ùë°[ùëõ]=y[n].ùë¶2[ùëõ].

s = y + y2 # tercera se√±al s[n] como la suma de y[n] y y2[n]
t = y * y2 # cuarta se√±al t[n] como el producto de y[n] y y2[n]
```

```python


# m) Graficar las se√±ales y[n] y y2[n]
plt.figure (1)
plt.plot(n, y, color='blue', label='y[n] = sin(œÄ * 0.12n)')
plt.plot(n, y2, color='green', label='y2[n] = cos(2œÄ * 0.03n)')
plt.title('Gr√°fica de y[n] y y2[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()

# Graficar las se√±ales s[n] y t[n]
# si fuera discreta, plt.stem
plt.figure (2)
plt.plot(n, s, color='blue', label='s[n] = y[n] + y2[n]')
plt.plot(n, t, color='red', label='t[n] = y[n] * y2[n]')
plt.title('Gr√°fica de s[n] y t[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
def reto(dic):
  arrayn=np.array([])
  for i in dic:
    arrayn=np.append(arrayn,dic[i])
  mean=np.mean(arrayn)
  max=np.max(arrayn)
  min=np.min(arrayn)
  std=np.std(arrayn)
  return(pd.Series([mean,max,min,std],index=["media","max","min","std"]))


D = {'a':1, 'b':2, 'c':3}
print(reto(D))
```
